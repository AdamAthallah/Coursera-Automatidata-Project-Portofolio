```python
# How many of each payment type are represented in the data?
df.groupby('payment_type').size()
```

```python
#==> Counting the  average tip for trips paid for with credit card?

df.loc[df['payment_type']== 1, 'tip_amount'].mean()

#  ==> The average tip for trips paid with credit card is 2.7298
```

```python
# ==> Counting the average tip for trips paid for with cash?

df.loc[df['payment_type']== 2, 'tip_amount'].mean()

# ==> the driver are not getting tipped
```

```python
#==> Searching for how much each vendorID is represented in the data
df['VendorID'].value_counts()

# vendor 1 is 12626
# vendor 2 is 10073
```

```python
#==> Searching for mean total amount of each vendor

df.groupby('VendorID')['total_amount'].mean()

# both vendors are represented by their mean at a similar number :
# vendor 1: 16.298119
# vendor 2: 16.320382
```

```python
# Filter the data for credit card payments only
mask3 = df['payment_type'] == 1
df[mask3]
```

```python
# Filter the credit-card-only data for passenger count only
mask4 = (df['payment_type'] == 1) & (df['passenger_count'] > 0)
df[mask4]
```

```python
# Calculate the average tip amount for each passenger count (credit card payments only)
mask5 = (df['passenger_count'] >= 0) & (df['payment_type'] == 1)
df[mask5].groupby('passenger_count')['tip_amount'].mean()
```
